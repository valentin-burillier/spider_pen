```python
import vpype as vp
import __init__ as sp
import matplotlib.pyplot as plt
```

La bibliothèque `vpype` est un prérequis à cette utilisation du robot.

Il est possible de générer facilement des images au format svg à partir d'image jpg ou png grace à ce [convertisseur](https://convertio.co/fr/jpg-svg/).

```python
file = r'pictures/cat.svg'

paths = vp.read_svg(file, quantization=1)[0]
```

```python
p = sp.Plotter(lenght=820, height=1140)
```

```python
p.clear()
zero = paths[0][0]
for path in paths:
    path -= zero
    p.goto(path[0].real, -path[0].imag)
    p.down()
    for point in path[1:]:
        p.goto(point.real, -point.imag)
    p.up()
```

```python
p.show()
```

```python
p.connect(L_Gi=740, L_Di=700)
```

```python
p.home()
p.disconnect()
```
